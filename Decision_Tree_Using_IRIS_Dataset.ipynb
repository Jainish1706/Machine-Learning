```python
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn import tree
from sklearn import datasets
```

```python
irisset = datasets.load_iris()
X = irisset.data
Y = irisset.target
```

```python
df = sns.load_dataset('iris')
X = df.iloc[:,0:4]
Y = df.iloc[:,4]
```

```python
cf = DecisionTreeClassifier(random_state = 1234)
cf.fit(X,Y)
```

```python
decPlot = plot_tree(decision_tree = cf, feature_names = ["sepal_length","sepal_width","petal_length","petal_width"],class_names = ["setosa","vercicolor", "verginica"], filled = True, precision = 4, rounded = True)
```
