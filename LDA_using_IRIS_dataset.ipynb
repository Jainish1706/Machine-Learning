```python
from sklearn import datasets
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
irisset = datasets.load_iris()
X = irisset.data
Y = irisset.target
```

```python
lda = LinearDiscriminantAnalysis(n_components = 2)
Xl = lda.fit(X,Y).predict(X)
cmat = confusion_matrix(Y, Xl)
print(cmat)
ldascore = accuracy_score(Y,Xl)
print(ldascore)
```

```python
Xl = lda.transform(X)
# print(Xl)
plt.scatter(Xl[:,0], Xl[:,1], c = Y)
plt.suptitle('LDA Iris Data')
plt.xlabel('LDA1')
plt.ylabel('LDA2')
plt.grid(1, which = 'both')
plt.axis('tight')
plt.show()
```

```python
pca = PCA(n_components = 2)
Xp = pca.fit(X).transform(X)
```

```python
plt.scatter(Xp[:,0], Xp[:,1], c = Y)
plt.suptitle('PCA Iris Data')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid(1, which= 'both')
plt.axis('tight')
plt.show()
```
